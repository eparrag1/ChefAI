```python
import numpy as np
import pandas as pd
import re
```

```python
file1 = open('file.csv', 'r')
Lines = file1.readlines()
```

```python
Lines
```

```python
for i in range(len(Lines)):
    ingredients = Lines[i].split('/food/')[1:]
    for x in range(len(ingredients)):
        ingredients[x] = re.sub(r'\n', '', ingredients[x])
        ingredients[x] = re.sub(r',', '', ingredients[x])
    Lines[i] = ingredients
    
```

```python
full_inventory = [item for sublist in Lines for item in sublist]
inventory = np.unique(full_inventory)
inventory = [re.sub(',', '', x) for x in inventory]
len(inventory)
```

```python
pd.set_option('display.max_rows',1000)
counts = pd.DataFrame({'A':full_inventory})
counts.value_counts().reset_index()
```

```python
food_dict={}

for i in range(len(inventory)):
    food_dict[inventory[i]] = i
    
```

```python
food_dict
```

```python
food_matrix = np.zeros([len(inventory),len(inventory)])
food_matrix
```

```python
def add_to_matrix(val,matrix):
    sublist = [x for x in Lines if inventory[val] in x]
    for i in range(len(sublist)):
        list_ = np.unique(sublist[i])
        for x in list_:
            if x != inventory[val]:
                item = food_dict[x]
                matrix[val,item] = matrix[val,item]+1
    return(matrix)
```

```python
for i in range(len(inventory)):
    food_matrix = add_to_matrix(i,food_matrix)
    
```

```python
import collections
    
def find_matches(ingredient,n):
    x = food_dict[ingredient]
    array = food_matrix[x,:]
    keys = np.linspace(0,len(array),len(array)+1)
    dict_2 = {}
    for i in range(len(array)):
        dict_2[keys[i]] = array[i]
    count = []
    food = []
    for key, value in dict_2.items():
        if value != 0:
            count.append(value)
            for a,b in food_dict.items():
                if b == key:
                    food.append(a)
    
    df = pd.DataFrame({'Count':count,'Food':food})
    df = (df.sort_values('Count',ascending=False))
    return(df.head(n))
```

```python
find_matches('mutton',30)
```
