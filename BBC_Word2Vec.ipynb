```python
import gensim
from gensim.models import KeyedVectors
from gensim.models import phrases

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("preprocessed_layer1.csv")
data['ingredients'] = data['ingredients'].str.replace('/t','')
data['instructions'] = data['instructions'].str.replace('/t','')
data['ingredients']  = data['ingredients'].str.replace('[^\w\s]','')
data['instructions']  = data['instructions'].str.lower()
data['instructions']  = data['instructions'].str.replace('(','')
data['instructions']  = data['instructions'].str.replace(')','')
data['instructions']  = data['instructions'].str.replace(',','')
data['instructions']  = data['instructions'].str.replace('.','')
data['instructions']  = data['instructions'].str.replace(':','')
data['instructions']  = data['instructions'].str.replace(';','')
data['instructions']  = data['instructions'].str.replace('!','')
data['instructions']  = data['instructions'].str.replace('  ',' ')
data['instructions']  = data['instructions'].str.replace('&','')
data['instructions']  = data['instructions'].str.replace('/',' ')


data.head(10)
```

```python
import numpy as np
import pandas as pd
import re

file1 = open('file.csv', 'r',encoding='utf-8')
Lines = file1.readlines()

for i in range(len(Lines)):
    ingredients = Lines[i].split('/food/')[1:]
    for x in range(len(ingredients)):
        ingredients[x] = re.sub(r'\n', '', ingredients[x])
        ingredients[x] = re.sub(r',', '', ingredients[x])
    Lines[i] = ingredients
Lines = [i for item in Lines for i in item]
```

```python
Lines_multi_words = []
Lines_space_words = []

for i in Lines:
    if '_' in i:#
        Lines_multi_words.append(i)
        i = i.replace('_',' ')
        Lines_space_words.append(i)
        
print(Lines_multi_words[0:10])
print(Lines_space_words[0:10])
```

```python
data['instructions_multi']  = data['instructions'] 
```

```python
from gensim.models import Phrases

from gensim.models.phrases import Phraser

documents = list(data['instructions'])

sentence_stream = [doc.split(" ") for doc in documents]

bigram = Phrases(sentence_stream, min_count=1, threshold=2)

bigram_phraser = Phraser(bigram)
```

```python
' '.join(bigram_phraser[data['instructions'][18000].split()])
```

```python
for i in range(len(data)):
    data['instructions_multi'][i] = ' '.join(bigram_phraser[data['instructions'][i].split()])
```

```python
data.to_csv('recipes1m_multi.csv')
```

```python
from gensim.models import Word2Vec
texts = []
for i in range(len(data)):
    texts.append(data['instructions_multi'][i].split(' '))
def word2vec(data):  
    model = Word2Vec(sentences=texts,  window=3, min_count=1)
    model.init_sims(replace=True)
    return(model)
model = word2vec(data)
```

```python
model.save('Recipes_bigrams_W2V.model')
```
