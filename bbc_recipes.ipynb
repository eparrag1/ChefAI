```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy import Request
import numpy as np
import pandas as pd
import requests
import lxml
import line_profiler
import time
import csv
import warnings
import pandas as pd

warnings.filterwarnings('ignore')
```

```python
class SpiderSpider(scrapy.Spider):
    name = 'spider'
    handle_httpstatus_list = [404,429]
    
    def errback_httpbin(self, response):
        pass
    
    def __init__(self):
        self.allowed_domains = ['https://www.bbc.co.uk/food']
        self.start_urls=[]
        self.outfile = open("recipe_links.csv", "w", newline="")
        self.writer = csv.writer(self.outfile)
        for i in range(1,50):
            for j in ['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z','0-9']:
                self.start_urls.append(f"https://www.bbc.co.uk/food/recipes/a-z/{j}/{i}#featured-content")


    def closed(self,reason):
        self.outfile.close()

        
    def parse(self, response):
        print(response.request.url)
        link = response.xpath('//a[@class="promo promo__other"]/@href').getall()
        print(link)
        self.writer.writerow(link)
        link = response.xpath('//a[@class="promo promo__main_course"]/@href').getall()
        print(link)
        self.writer.writerow(link)
        link = response.xpath('//a[@class="promo promo__cakes_and_baking"]/@href').getall()
        print(link)
        self.writer.writerow(link)
    
```

```python
%%timeit -n 1 -r 1
from scrapy.crawler import Crawler
if __name__ == "__main__":
    process = CrawlerProcess({
        'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
    })
    
    process.crawl(SpiderSpider)
    process.start() 
    process.stop()
```
